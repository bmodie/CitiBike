```python
import pandas as pd
import numpy as np
```

```python
# Read in all 2017 rider data
january_2017 = pd.read_csv('201701-citibike-tripdata.csv')
february_2017 = pd.read_csv('201702-citibike-tripdata.csv')
march_2017 = pd.read_csv('201703-citibike-tripdata.csv')
april_2017 = pd.read_csv('201704-citibike-tripdata.csv')
may_2017 = pd.read_csv('201705-citibike-tripdata.csv')
june_2017 = pd.read_csv('201706-citibike-tripdata.csv')
july_2017 = pd.read_csv('201707-citibike-tripdata.csv')
august_2017 = pd.read_csv('201708-citibike-tripdata.csv')
september_2017 = pd.read_csv('201709-citibike-tripdata.csv')
october_2017 = pd.read_csv('201710-citibike-tripdata.csv')
november_2017 = pd.read_csv('201711-citibike-tripdata.csv')
december_2017 = pd.read_csv('201712-citibike-tripdata.csv')
```

```python
# Merge the 2017 monthly data and sort by trip_id
citibike_2017 = pd.concat([january_2017, february_2017, march_2017, april_2017, may_2017,
                       june_2017, july_2017, august_2017, september_2017,
                       october_2017, november_2017, december_2017])
citibike_2017 = citibike_2017.sort_values(by='Start Time')
```

```python
# Save the DF to csv for use in Tableau
citibike_2017.to_csv("citibike_2017.csv", index=False, header=True)
```
